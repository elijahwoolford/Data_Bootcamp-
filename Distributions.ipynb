```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
# Binomial distribution 

binomial = np.random.binomial(1,.5,100)
plt.hist(binomial)
plt.axvline(binomial.mean(), color='b', linestyle='solid', linewidth=2)
plt.axvline(binomial.std(), color='b', linestyle='dashed', linewidth=2)
plt.show()
```

The mean and standard deviation have no bearing on the above distribution. 

```python
# Poisson distribution 

poisson = np.random.poisson(3,100)
plt.hist(poisson)
plt.axvline(poisson.mean(), color='b', linestyle='solid', linewidth=2)
plt.axvline(poisson.std(), color='b', linestyle='dashed', linewidth=2)
plt.show()
```

This distribution shows some skew to the right of the mean.  

```python
# Exponential distribution 

exponential = np.random.exponential(1,100)
plt.hist(exponential)
plt.axvline(exponential.mean(), color='b', linestyle='solid', linewidth=2)
plt.axvline(exponential.std(), color='b', linestyle='dashed', linewidth=2)
plt.show()
```

This distribution has most of its data very close to the mean and standard deviation. Also most of the data occurs at the low end of the spectrum

```python
# Gamma distribution

gamma = np.random.gamma(5,1,100)
plt.hist(gamma)
plt.axvline(gamma.mean(), color='b', linestyle="solid", linewidth=2)
plt.axvline(gamma.std(), color='b', linestyle='dashed', linewidth=2)
plt.show()
```

Similar to the normal distribution, gamma produces a mean close to the center and a slightly higher standard deviation due to the presence of an outlier.

```python
# Normal distribution

normal = np.random.normal(1,.5,100)
plt.hist(normal)
plt.axvline(normal.mean(), color='b', linestyle='solid', linewidth=2)
plt.axvline(normal.std(), color='b', linestyle='dashed', linewidth=2)
plt.show()
```

As expected the normal distribution has the mean close to the center and a low standard deviation.

```python
# Logistic distribution

logistic = np.random.logistic(5,1,100)
plt.hist(logistic)
plt.axvline(logistic.mean(), color='b', linestyle='solid', linewidth=2)
plt.axvline(logistic.std(), color='b', linestyle='dashed', linewidth=2)
plt.show()
```

Similar to a normal distribution.

```python
# Comparison of two normal distributions

norm1 = np.random.normal(5,.5,1000)
norm2 = np.random.normal(10,1,1000)
norm3 = norm1 + norm2
plt.hist(norm3)
plt.axvline(norm3.mean(), color='b', linestyle='solid', linewidth=2)
plt.axvline(norm3.mean()+norm3.std(), color='b', linestyle='dashed', linewidth=2)
plt.axvline(norm3.mean()-norm3.std(), color='b', linestyle='dashed', linewidth=2)
plt.show()
```

Adding two distributions together result in the new distrubutions having a mean of the means of the distributions added together.
